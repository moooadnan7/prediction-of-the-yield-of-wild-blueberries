```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
data=pd.read_csv("train.csv")
```

```python
x=data.drop("yield",axis=1)
x
```

```python
y=data["yield"]
y
```

```python
model=LinearRegression()
model.fit(x,y)
```

```python
model.predict([[14996,290.0,12.5,0.25,0.25,0.63,0.75,69.7,42.1,58.2,50.2,24.3,41.2,16.0,0.26,0.520637,0.419216,35.466874]])
```

```python
model.score(x,y)
```

```python
d_test=pd.read_csv("test.csv")
d_test
```

```python
model.predict(d_test)
```

```python
d_test["yield"]=model.predict(d_test)
d_test
```
